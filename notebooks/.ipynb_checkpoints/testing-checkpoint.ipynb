```python
import numpy as np
import scipy as sp
import scipy.signal as signal
import pickle as pickle
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ladder = pickle.load(open('../../ladder.pkl', 'rb'))
plt.imshow(ladder, cmap='Greys')
his = ladder.mean(axis=1)
```

```python
plt.plot(his)
```

```python
filt = his.copy()
filt[filt < 0.4] = 0
```

```python
plt.plot(filt)
ax = plt.gca()
ax.set_xticklabels('bob', rotation=90)
```

```python
peaks = signal.find_peaks_cwt(filt, widths=np.arange(3, 5))
```

```python
weights = ['1500', '1000', '700', '500', '400', '300', '200', '100', '50', '25']
```

```python
locs = dict(zip(peaks, weights))
```

```python
fig, ax = plt.subplots(1,1, figsize=(10, 10))
ax.imshow(ladder, cmap='Greys')
ax.yaxis.tick_right()
ax.set_yticks(peaks)
ax.set_yticklabels(weights)
```
